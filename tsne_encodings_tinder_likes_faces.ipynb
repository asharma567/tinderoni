```python
from tsne import multithread_map, rescaler
import os, glob
from tsne import image_scatter_plot
from PIL import Image
from tsne_grid_search_tool import find_best_tsne_plot, save_plots_with_params
from sklearn.manifold import TSNE
import pickle
import numpy as np
```

```python
data = pickle.loads(open('encodings.pickle', "rb").read())
paths = [d['imagePath']  for d in data]
data = np.array(data)
encodings = [d["encoding"] for d in data]

arrays_rescaled = list(map(rescaler, paths))


X = np.array(list(map(lambda x: np.array(np.array(x, dtype=np.float32)), arrays_rescaled)))
```

```python
params_dic = find_best_tsne_plot(encodings)
save_plots_with_params(params_dic, 'encodings_tinder_likes_faces')
```

```python
#open the image plots here
```

```python
params = {
#     'init':'pca',
	'perplexity': 10,
	'n_iter': 5000,
	'learning_rate': 1000,
	'early_exaggeration': 6.0,
    'random_state': 1
}

tsne_embeddings_lis = [TSNE(**params).fit_transform(encodings) for _ in range(5)]
```

```python
for embeddings in tsne_embeddings_lis:

    plot_data = image_scatter_plot(
                    tsne_scatter_plot=embeddings, 
                    images=X, 
                    res=500
                )

    print(plot_data.shape)
    im = Image.fromarray(np.uint8(plot_data))
    im.show()
```
