```python
import pandas as pd
file_name = 'SCUT-FBP5500_v2/All_Ratings.xlsx'
df = pd.read_excel(file_name)
```

```python

paths = '/Users/ajay/Desktop/project_prototypes/tinderoni/SCUT-FBP5500_v2/Images/'
```

```python
lis = df[(df['Filename'].str.startswith('CF')) & (df['Rating'] == 1 )]['Filename'].tolist()
hot = [paths + img for img in lis]
```

```python
lis = df[(df['Filename'].str.startswith('CF')) & (df['Rating'] == 5 )]['Filename'].tolist()
not_hot = [paths + img for img in lis]
```

```python
main(
    path=hot, 
    save_fname='CF_superhot.jpeg',
    save_plots_fname='CF_superhot'
)
```

```python
main(
    path=not_hot, 
    save_fname='CF_supernot.jpeg',
    save_plots_fname='CF_supernot'
)
```
