comparing the various data dispersion metrics. Sometimes you want them robust to outliers and sometimes you want it to capture the outliers and be greatly sensitive to it. since std uses mean squared error, it'll have the highest sensitivity to it.


https://stackoverflow.com/questions/8930370/where-can-i-find-mad-mean-absolute-deviation-in-scipy


```python
from numpy import mean, absolute, std, median

def mad(data, axis=None):
    return mean(absolute(data - mean(data, axis)), axis)

def mad2(data, axis=None):
    return median(absolute(data - median(data, axis)), axis)

```

```python
from statsmodels import robust
```

```python

sample2 = [0.4, 0.7, 0.1]
sample3 = [0.1, 0.1, 0.9]
sample4 = [0.1, 0.1, 0.5]
sample5 = [0.1, 0.9, 0.3]
```

```python
robust.mad([ 66, 69, 76, 72, 79, 77, 74, 77, 71, 79, 74, 66, 67, 73 ], axis=0)
```

```python
std(sample2), mad(sample2), mad2(sample2)
```

```python
print(std(sample3), mad(sample3), mad2(sample3))
print(std(sample4), mad(sample4), mad2(sample4))
print(std(sample5), mad(sample5), mad2(sample5))
```
