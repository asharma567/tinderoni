```python
from modeling_helpers import *
from xgboost import XGBClassifier
```

```python
tinder_labels = np.r_[np.zeros(num_of_dislikes), np.ones(len(paths) - num_of_dislikes )]
```

```python
model = XGBClassifier(learning_rate=0.02, n_estimators=600, objective='binary:logistic',
                        silent=True, nthread=1)
```

```python
random_grid_searching_model(model, np.array(encodings), tinder_labels,params=full_param_grid, folds=10)
```

```python
positives = sum(tinder_labels)
negs = tinder_labels.shape[0]

positives/negs, negs/positives
```

```python
full_param_grid
```

```python
partial_param_grid = {
    'subsample': [0.8],
    'scale_pos_weight': [1, positives/negs, negs/positives],
    'n_estimators': [1000,1500],
    'min_child_weight': [10],
    'max_depth': [4],
    'gamma': [1.5],
    'colsample_bytree': [0.3,0.6]
}
```

```python
results = full_gridsearch(model, np.array(encodings), tinder_labels,param_grid=partial_param_grid, folds=10)
results
```
